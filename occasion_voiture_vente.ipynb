```python
#histogramme de distribution de prix de vente
plt.figure(figsize=(8,5))
plt.hist(df['selling_price'],color = 'magenta', edgecolor= 'yellow', bins =40)
plt.title('distribution de prix de vente')
plt.xlabel('prix de vente')
plt.ylabel('nombre de véhicule')
plt.grid()
plt.savefig('mon_graphique1',dpi=400)
plt.show()
```

```python
#Distribution de l'année de fabrication de véhicules
plt.figure(figsize=(8,5))
plt.hist(df['year'], color = 'cyan', edgecolor = 'green', bins =20)
plt.title("distribution de l'année de fabrication")
plt.xlabel('Année')
plt.ylabel('nombre ')
plt.grid()
plt.savefig('mon_graphique1',dpi=400)
plt.show()
```

```python
#distribution de kilomètre parcouri
plt.figure(figsize=(10,8))
plt.hist(df['km_driven'], color = 'blue', edgecolor= 'cyan', bins= 10)
plt.title('distribution de kilomètre parcouri')
plt.xlabel('Kilomètre parcouri')
plt.ylabel('nombre de véhicule')
plt.grid()
plt.savefig('mon_graphique2', dpi = 400)
plt.show()
```

```python
#calucul des moyenne de prix de vente des véhicules
df_mean_price = df_drop_doublon.groupby('fuel')['selling_price'].mean().reset_index()
df_mean_price
```

```python
# prix moyen  par type de carburant
plt.figure(figsize=(10,5))
sns.barplot(data=df_mean_price, x= 'fuel', y = 'selling_price')
plt.title('Prix moyen par type de carburant')
plt.xlabel('carburant')
plt.ylabel('moyenne')
plt.grid()
plt.savefig('mon_graphique4', dpi= 300)
plt.show()
```
